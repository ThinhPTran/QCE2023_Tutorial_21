# Implementation of an n-qubit Quantum Fourier Transform Gate

image.png

```python
# Number of Qubit
n = 3
# Size of the Hilbert Space/ Matrix
N = pow(2,n)
QFT = np.zeros((N,N), dtype = 'complex_') # need complex to store complex number
for k in range(N):
  for l in range(N):
    QFT[k,l] = 1/np.sqrt(N)*np.exp(-(1j)*k*l*2*np.pi/N)

np.set_printoptions(precision=3)
print(QFT)
```
